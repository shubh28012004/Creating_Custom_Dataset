```python
#List of 6 features
features = ["id" , "gender" , "subscriber" , "name" , "rating" , "mail_id"]
```

```python
#creating a dataframe for these features
df = pd.DataFrame(columns=features)
df
```

**Creating Imbalanced Data**

Generate unique identifiers

```python
num_users=100
df['id']=[uuid.uuid4().hex for i in range(num_users)]
df
```

```python
df['id']=[uuid.uuid4().int for i in range(num_users)]
df
```

to check whether ids are unique or not

```python
print(df['id'].nunique()==num_users)
```

Generating Gender values

```python
genders = ["mail","female","na"]
df['gender'] = random.choices(genders , weights = (45,45,10),k=num_users) #k is a parameter of num.choices
df
```

Generating Subscriber values

```python
choice = [True , False]
df['subscriber']=random.choices(choice , k=num_users)
df
```

Generating name values

```python
#instantitaing faker
faker = Faker(locale = "EN_IN")
def name_gen(gender):
  if gender == 'male':
    return faker.name_male()
  elif gender == 'female':
    return faker.name_female()
  return faker.name()

#Generating names
df['name'] = [name_gen(i) for i in df['gender']]
df
```

Generating rating values

```python
rating = [1,2,3,4,5]
df['rating'] = random.choices(rating , weights = (45,45,20,30,10), k = num_users)
df
```

Generating mail ids using faker

```python
import numpy as np
from faker import Faker
import pandas as pd

# Create a Faker object
fake = Faker()

def faker_categorical(gender, seed=None):
    np.random.seed(seed)
    fake.seed_instance(seed)
    gender = np.random.choice(["M", "F"], p=[0.5, 0.5])
    return {
        "First name": fake.first_name_male() if gender == "M" else fake.first_name_female(),
        "Last name": fake.last_name(),
        "E-mail": fake.ascii_email(),
    }

# Sample DataFrame with a 'gender' column
data = {
    "gender": ["M", "F", "M", "F", "M"]
}
df = pd.DataFrame(data)

# Use the faker_categorical function to populate the 'mail_id' column
df['mail_id'] = df['gender'].apply(lambda gender: faker_categorical(gender))

# If you want to access individual fields like 'First name', 'Last name', and 'E-mail', you can use DataFrame apply
df['First name'] = df['mail_id'].apply(lambda x: x['First name'])
df['Last name'] = df['mail_id'].apply(lambda x: x['Last name'])
df['E-mail'] = df['mail_id'].apply(lambda x: x['E-mail'])

# Drop the 'mail_id' column if you don't need it anymore
df.drop(columns=['mail_id'], inplace=True)

print(df)
```

Convert dataframe to dataset

```python
df.to_csv('dpldataset.csv')
```
